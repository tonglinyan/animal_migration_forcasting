```python
lat_min = min(data['Latitude'])
lat_max = max(data['Latitude'])
long_min = min(data['Longitude'])
long_max = max(data['Longitude'])
```

```python
plt.figure(figsize=(16,8))   
map = Basemap(llcrnrlon = lat_min-1, llcrnrlat = long_min-1, urcrnrlon = lat_max+1, urcrnrlat = long_max+1, resolution = 'h')
map.drawmapboundary(fill_color = 'aqua')
map.fillcontinents(color = 'gray', lake_color = 'aqua')
map.drawcoastlines()
plt.show()
```
