# Import data

```python
def to_list(data):
    return_list = []
    for name, columns in data.iteritems():
        return_list.extend(columns.values.tolist())
    return_list = list(map(int, return_list))
    return return_list
```

```python
def accuracy_plot(true, pred):
    plt.scatter(to_list(true), to_list(pred), s = 10, c = to_list(-true), cmap='viridis')  
    plt.ylabel('predicted migration intensity(birds km-3)')
    plt.ylabel('actual migration intensity(birds km-3)')
    x = np.arange(0, 160000)
    plt.fill_between(x, x*0.9, x*1.1, alpha = 0.3, where=None, interpolate=False, step=None, data=None)
    plt.show()
```

```python
y_pred_ARIMA = pd.read_csv('../dataset/radar/pred_ARIMA.csv')
y_pred_log = pd.read_csv('../dataset/radar/pred_log.csv')
y_pred_RF1 = pd.read_csv('../dataset/radar/pred_RF1.csv')
y_pred_RF2 = pd.read_csv('../dataset/radar/pred_RF2.csv')
y_pred_RF3 = pd.read_csv('../dataset/radar/pred_RF3.csv')
y_pred_RF4 = pd.read_csv('../dataset/radar/pred_RF4.csv')
y_pred_RF5 = pd.read_csv('../dataset/radar/pred_RF5.csv')
y_pred_RF6 = pd.read_csv('../dataset/radar/pred_RF6.csv')
y_pred_RF7 = pd.read_csv('../dataset/radar/pred_RF7.csv')
y_pred_RF8 = pd.read_csv('../dataset/radar/pred_RF8.csv')
y_pred_RF9 = pd.read_csv('../dataset/radar/pred_RF9.csv')
data = pd.read_csv('../dataset/radar/time_filled_AR.csv')
data = data.drop(['samplingperiod'], axis = 1)

y_pred_ARIMA = y_pred_ARIMA[data.columns]
y_pred_log = y_pred_log[data.columns]
y_pred_RF1 = y_pred_RF1[data.columns]
y_pred_RF2 = y_pred_RF2[data.columns]
y_pred_RF3 = y_pred_RF3[data.columns]
y_pred_RF4 = y_pred_RF4[data.columns]
y_pred_RF5 = y_pred_RF5[data.columns]
y_pred_RF6 = y_pred_RF6[data.columns]
y_pred_RF7 = y_pred_RF7[data.columns]
#y_pred_RF8 = y_pred_RF8[data.columns]
y_pred_RF9 = y_pred_RF9[data.columns]
#print(len(y_pred_RF_u))
#print(data[len(data)-len(y_pred_RF_u):])
```

# Accuracy plot

```python
accuracy_plot(data[len(data)-len(y_pred_RF1):], y_pred_RF1)
```

# Mean absolute error

```python
from sklearn.metrics import mean_absolute_error
print('Mean absolute error with ARIMA model: %f'%mean_absolute_error(data[len(data)-len(y_pred_ARIMA):], y_pred_ARIMA))
print('Mean absolute error with ARIMA model (log): %f'%mean_absolute_error(data[len(data)-len(y_pred_log):], y_pred_log))
print('Mean absolute error with Random Forest 1: %f'%mean_absolute_error(data[len(data)-len(y_pred_RF1):], y_pred_RF1))
print('Mean absolute error with Random Forest 2: %f'%mean_absolute_error(data[len(data)-len(y_pred_RF2):], y_pred_RF2))
print('Mean absolute error with Random Forest 3: %f'%mean_absolute_error(data[len(data)-len(y_pred_RF3):], y_pred_RF3))
print('Mean absolute error with Random Forest 4: %f'%mean_absolute_error(data[len(data)-len(y_pred_RF4):], y_pred_RF4))
print('Mean absolute error with Random Forest 5: %f'%mean_absolute_error(data[len(data)-len(y_pred_RF5):], y_pred_RF5))
print('Mean absolute error with Random Forest 6: %f'%mean_absolute_error(data[len(data)-len(y_pred_RF6):], y_pred_RF6))
print('Mean absolute error with Random Forest 7: %f'%mean_absolute_error(data[len(data)-len(y_pred_RF7):], y_pred_RF7))
print('Mean absolute error with Random Forest 8: %f'%mean_absolute_error(data[len(data)-len(y_pred_RF8):], y_pred_RF8))
print('Mean absolute error with Random Forest 9: %f'%mean_absolute_error(data[len(data)-len(y_pred_RF9):], y_pred_RF9))
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score
def evaluation(pred, true)
    f1 = f1_score(true, pred, average='macro' )
    p = precision_score(true, pred, average='macro')
    r = recall_score(true, pred, average='macro')
    return f1, p, r
print(evaluation(to_list(data[len(data)-len(y_pred_RF1):]), to_list(y_pred_RF1)))
```
