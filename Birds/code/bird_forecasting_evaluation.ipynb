```python
def to_list(data):
    return_list = []
    for name, columns in data.iteritems():
        return_list.extend(columns.values.tolist())
    return_list = list(map(int, return_list))
    return return_list
```

```python
def accuracy_plot(true, pred):
    plt.scatter(to_list(true), to_list(pred), s = 10, c = to_list(-true), cmap='viridis')  
    plt.ylabel('predicted migration intensity(birds km-3)')
    plt.ylabel('actual migration intensity(birds km-3)')
    x = np.arange(0, 160000)
    plt.fill_between(x, x*0.9, x*1.1, alpha = 0.3, where=None, interpolate=False, step=None, data=None)
    plt.show()
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score
def evaluation(pred, true):
    f1 = f1_score(true, pred, average='macro' )
    p = precision_score(true, pred, average='macro')
    r = recall_score(true, pred, average='macro')
    return f1, p, r

print(evaluation(to_list(data[len(data)-len(y_pred_RF1):]), to_list(y_pred_RF1)))
print(evaluation(to_list(data[len(data)-len(y_pred_RF2):]), to_list(y_pred_RF2)))
print(evaluation(to_list(data[len(data)-len(y_pred_RF3):]), to_list(y_pred_RF3)))
print(evaluation(to_list(data[len(data)-len(y_pred_RF4):]), to_list(y_pred_RF4)))
print(evaluation(to_list(data[len(data)-len(y_pred_RF5):]), to_list(y_pred_RF5)))
print(evaluation(to_list(data[len(data)-len(y_pred_RF6):]), to_list(y_pred_RF6)))
```

# R2 plot

```python
def map_plot(data, lat_min, lat_max, lon_min, lon_max, name):
    data.dropna()
    plt.figure()
    earth = Basemap(llcrnrlon=lon_min, llcrnrlat=lat_min, urcrnrlon=lon_max, urcrnrlat=lat_max)
    earth.drawcountries()
    earth.bluemarble(alpha=0.42)
    earth.drawcoastlines(color='#555566', linewidth=1)
    cm = plt.cm.get_cmap('Accent')
    earth.scatter(data['lon'], data['lat'],c = data.iloc[:,2],cmap = 'rainbow')
    plt.colorbar()
    plt.xlabel("longitude")
    plt.ylabel("latitude")
    plt.savefig('../figures/%s.png'%name, dpi=350)
    plt.show()
```

```python
data = pd.read_csv('../dataset/radar_narr_combine.csv')
result = pd.read_csv('../dataset/result_mean_estimator.csv')
lat_lon = data[['radar_id', 'lat', 'lon']]
lat_lon.drop_duplicates(['radar_id', 'lat', 'lon'], keep = 'first', inplace = True)
lat_lon.set_index('radar_id', inplace = True)
result.set_index('radar', inplace = True)
lat_lon['r2'] = result['r2']
lat_lon['mse'] = result['mse']
loc = lat_lon['r2'][lat_lon['r2']<0].index.tolist()
for i in loc:
    lat_lon.loc[i, 'r2'] = 0
```

```python
lat_min = min(lat_lon['lat'])
lat_max = max(lat_lon['lat'])
lon_min = min(lat_lon['lon'])
lon_max = max(lat_lon['lon'])
```

```python
map_plot(lat_lon[['lat', 'lon', 'r2']], lat_min-2, lat_max+2, lon_min-2, lon_max+2, 'r2')
map_plot(lat_lon[['lat', 'lon', 'mse']], lat_min-2, lat_max+2, lon_min-2, lon_max+2, 'mse')
```
